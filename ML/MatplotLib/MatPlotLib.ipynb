```python
import pandas as pd;
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("sharma-kohli.csv")
```

```python
df
```

```python
plt.plot(df["RG Sharma"])
plt.plot(df["V Kohli"])
```

```python
plt.plot(df["RG Sharma"])
plt.plot(df["V Kohli"])
plt.title("Rohit Sharma Vs Virat Kohli Carreer")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

```python
plt.plot(df['index'],df["RG Sharma"],color="Black",label="Rohit")
plt.plot(df['index'],df["V Kohli"],color="yellow",label="Virat", marker='o')
plt.title("Rohit Sharma Vs Virat Kohli Carreer")
plt.xlabel("Season")
plt.ylabel("Run Scored")
plt.legend() 
plt.grid()
```
