```python
df=pd.DataFrame()
```

```python
df
```

```python
for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=4acc4fee9aca01b7721b3063d2381bd0&language=en-US&page={i}')
    
    # Check if the response is successful
    if response.status_code == 200:
        data = response.json()
        # Convert the 'results' field in the JSON response to a DataFrame
        newDf = pd.DataFrame(data.get('results', []))
        # Append the new data to the main DataFrame
        df = pd.concat([df, newDf], ignore_index=True)
    else:
        print(f"Failed to fetch data for page {i}. Status code: {response.status_code}")
```

```python
df.to_csv("Movies.csv")
```
