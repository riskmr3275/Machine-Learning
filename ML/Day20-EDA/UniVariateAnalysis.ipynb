```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head(6)
```

#  1. Categorical Data

## a Countplot

```python
sns.countplot(df["Embarked"],palette="pastel",legend=False)
# Rotate the x-axis labels (if necessary)
plt.xticks(rotation=1)  # 'horizontal' is the default
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.title("Count of Embarked Categories")
 
plt.show()
```

## b. PieChart

```python
df['Sex'].value_counts().plot(kind="pie",autopct="%.2f")
```

#### 2.Numerical Data

#### a. Histogram

```python
plt.hist(df['Age'],bins=10)
plt.show()
```

#### b.Distplot

```python
sns.distplot(df['Age'])
```

#### c.Boxplot

```python
sns.boxplot(df['Age'],vert=True)
```
